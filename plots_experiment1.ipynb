```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
exp1search1n = [70, 102, 152, 207, 266, 321, 376, 453, 494, 550]
exp1search1nstd = [18, 10, 11, 12, 11, 12, 14, 109, 17, 19]
df1n = pd.DataFrame(exp1search1n, columns=["responsetime"])
df1n['std'] = np.array(exp1search1nstd)


exp1search4n = [33, 52, 77, 104, 131, 160, 188, 216, 244, 273]
exp1search4nstd = [23, 6, 9, 7, 8, 8, 7, 7, 8, 8]
df4n = pd.DataFrame(exp1search4n, columns=["responsetime"])
df4n['std'] = np.array(exp1search4nstd)

exp1search8n = [36, 52, 77, 105, 131, 159, 186, 214, 246, 274]
exp1search8nstd = [21, 6, 6, 6, 8, 7, 6, 6, 7, 5]
df8n = pd.DataFrame(exp1search8n, columns=["responsetime"])
df8n['std'] = np.array(exp1search8nstd)
```

```python
fig1, ax1 = plt.subplots(figsize=(20, 10))
ax1.errorbar(x=[queries*50000 for queries in range(0, 10)], y=df1n['responsetime'], yerr=df1n['std'], label="Search time 1 node")
ax1.errorbar(x=[queries*50000 for queries in range(0, 10)], y=df4n['responsetime'], yerr=df1n['std'], label="Search time 4 nodes")
ax1.errorbar(x=[queries*50000 for queries in range(0, 10)], y=df8n['responsetime'], yerr=df1n['std'], label="Search time 8 nodes")

ax1.set_xticks([queries*50000 for queries in range(0, 10)])
ax1.set_title('Experiment 1: Search query response times ')
ax1.set_ylabel('Response time (ms)')
ax1.set_xlabel('Number of search queries')

ax1.legend(loc='upper left')
```
